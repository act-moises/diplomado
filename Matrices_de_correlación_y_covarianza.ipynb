```python
import pandas as pd

# Leer el archivo CSV
DataP2 = "P2_DB.csv"
df = pd.read_csv(DataP2)

print(df.head())
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Matriz de correlación de la data original
correlation_matrix = df[['V1', 'V2', 'V3','V4','V5', 'V6', 'V7','V8', 'V9', 'V10' ]].corr()
print("Matriz de correlación:\n", correlation_matrix)

# Visualización de la matriz de correlación
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```

```python
#Matriz de covarianzas
covariance_matrix = df[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10']].cov()
print("Matriz de varianzas y covarianzas:\n", covariance_matrix)

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 10))

# Crear el heatmap con formato ajustado
sns.heatmap(
    covariance_matrix,
    annot=True,            # Mostrar los valores en el heatmap
    fmt=".2f",             # Mostrar valores con dos decimales
    cmap='coolwarm',       # Cambiar la paleta de colores
    vmin=-10000, vmax=10000,  # Ajustar rango de valores para mejor contraste (ajústalo según tus datos)
    linewidths=0.5         # Agregar líneas entre celdas
)

# Agregar título y etiquetas
plt.title('Matriz de varianzas y covarianzas', fontsize=16)
plt.xticks(rotation=45, fontsize=12)  # Rotar etiquetas de columnas
plt.yticks(fontsize=12)               # Ajustar tamaño de etiquetas de filas
plt.tight_layout()                    # Ajustar el espaciado

# Mostrar la gráfica
plt.show()
```

```python
# Data normalizada
DataP3 = "P2_DBNCM.csv"
df1 = pd.read_csv(DataP3)

# Mostrar las primeras filas del DataFrame
print(df1.head())
```

```python
# Matriz de correlación de la data normalizada
correlation_matrix = df1[['V1', 'V2', 'V3','V4','V5', 'V6', 'V7','V8', 'V9', 'V10' ]].corr()
print("Matriz de correlación:\n", correlation_matrix)

# Visualización de la matriz de correlación
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```

```python
# Calcular la matriz de covarianzas de la data NORMALIZADA
covariance_matrix = df1[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10']].cov()
print("Matriz de varianzas y covarianzas:\n", covariance_matrix)

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 10))

# Crear el heatmap con formato ajustado
sns.heatmap(
    covariance_matrix,
    annot=True,            # Mostrar los valores en el heatmap
    fmt=".2f",             # Mostrar valores con dos decimales
    cmap='coolwarm',       # Cambiar la paleta de colores
    vmin=-0.2, vmax=0.2,  # Ajustar rango de valores para mejor contraste (ajústalo según tus datos)
    linewidths=0.5         # Agregar líneas entre celdas
)

# Agregar título y etiquetas
plt.title('Matriz de varianzas y covarianzas con datos normalizados', fontsize=16)
plt.xticks(rotation=45, fontsize=12)  # Rotar etiquetas de columnas
plt.yticks(fontsize=12)               # Ajustar tamaño de etiquetas de filas
plt.tight_layout()                    # Ajustar el espaciado

# Mostrar la gráfica
plt.show()
```
